# A Standards-Based Explorer for Environmental Data

*Finding, understanding, and visualizing data through its metadata*

- Search a data catalogue by **where**, **when**, and **what** — then view the data in the browser.
- Built entirely on **open standards**, not custom formats.

A small tool with a big idea — if data is described well, software can find it and show it automatically, with no special handling per dataset.

<span class="kicker">Motivation</span>

## We have lots of data. Using it is the hard part.

- Environmental data lives in many archives, formats, and projections.
- Users don't want files — they want to **find the right thing** and **look at it**.
- The gap between "data exists" and "data is usable" is **metadata**.

The bottleneck isn't storage — it's discovery and interpretation. Metadata is the bridge, and this project tests how well that bridge holds.

<span class="kicker">What it does</span>

## Search → understand → visualize, all from the catalogue

- **Search** by area (draw a box on a map), time range, and free text.
- The tool reads each result's **metadata** to decide what it is and how to show it.
- **Visualize**: plot a time series or profile inline, or overlay a map layer.

The user never touches a file path or a format. They draw a box, pick a date, and the catalogue's metadata does the rest.

<span class="kicker">Foundations</span>

## Built on open standards — it speaks the community’s languages

| Standard | Role here |
|---|---|
| **CSW** | Catalogue Service for the Web — search the catalogue |
| **ISO 19115 / 19139** | The metadata records themselves |
| **OGC WMS** | Ready-made map layers |
| **OPeNDAP / THREDDS** | Direct access to the underlying data |
| **CF Conventions** | Tells us the *kind* of measurement |

Every row is an international open standard. Because we don't invent formats, this works with any compliant catalogue — not just ours.

<span class="kicker">The key idea</span>

## The metadata decides the experience

- **CF "featureType"** (time series, profile, trajectory…) tells the tool **how** to plot the data — automatically.
- **Service links** in the record (WMS, OPeNDAP) tell it **how** to access the data.
- Result: **no per-dataset programming.** Good metadata = instant usability.

A field called "featureType" literally chooses the right chart. When metadata is rich, the tool just works. When it isn’t, the tool can’t help — a powerful diagnostic.

<span class="kicker">What makes it unique</span>

## From data to interactive product — in zero steps

- A standards-compliant dataset is **already** a product: point at its link, and the app builds the controls and the plot itself.
- No download. No format wrangling. No bespoke plotting code. No custom interface.

### Data to science — served on a silver plate.

This is the pitch. Everywhere else, turning a dataset into an interactive product is days of bespoke work — download, parse, write plotting code, build a UI. Here it is a single link, because the metadata already says everything the software needs.

<span class="kicker">The data model</span>

## One convention-driven pipeline — not custom code per dataset

**Data link** (OPeNDAP / NetCDF) → **open as a standard dataset** (xarray) → read **`featureType`** (CF convention) → **the right view, automatically** (time series · profile · trajectory)

- Variables, axes, labels and units come straight from **CF metadata**; quality-flag variables are auto-excluded.
- Add a compliant dataset → get a working, interactive product. **No per-dataset programming.**

Here is why "zero steps" is real, not a slogan. One pipeline: validate the link, open the data, read the featureType, and the convention tells us which interactive view to build. The dataset’s own metadata drives everything — variables, axes, labels, even which charts make sense.

<span class="kicker">Strengths</span>

## What's genuinely good

- ✅ **Interoperable** — open standards, no vendor or format lock-in.
- ✅ **Metadata-driven** — behavior comes from the catalogue, not hard-coded.
- ✅ **No data duplication** — reads existing archives in place.
- ✅ **Polar-ready** — supports Arctic projections (Svalbard, UPS North) most web maps ignore.
- ✅ **Lightweight** — runs in a browser, already live in our infrastructure.

<span class="kicker">Weaknesses &amp; hard lessons</span>

## Honest limitations — most are about data, not code

- ⚠️ **Only as good as the metadata.** Missing featureType or service links = invisible to the tool.
- ⚠️ **Catalogue noise.** Millions of fine-grained satellite records make useful datasets sparse and searches heavy.
- ⚠️ **Inconsistent catalogues.** One returned *different results for the same query* — a server-side compliance gap.
- ⚠️ **Projection fragmentation.** Services advertise different coordinate systems; polar data needs care.
- ⚠️ **Performance depends on remote servers**, not on us.

The most interesting finding: most of our problems were metadata-quality problems, not software problems. The tool is a mirror held up to our data management.

<span class="kicker">Response</span>

## Turning weaknesses into guardrails

- Guard against runaway queries — require a filter; warn when a search is too broad.
- Search results several times faster by querying the catalogue **in parallel**.
- Bundled our own **Norwegian + Svalbard coastline** where no standard polar basemap existed.
- A **layer manager** to control map overlays (order, transparency, visibility).

Each limitation pushed a concrete improvement — several are reusable building blocks for other tools.

<span class="kicker">Status</span>

## Where we are today

- 🟢 **Usable working prototype**, deployed in our operational environment.
- 🟢 Refactored into **reusable, shared components** (search, plotting, maps).
- 🟢 **Automated test suite** (130+ tests) for confidence in changes.
- 🟡 Validated mainly against MET Norway catalogues — broader testing pending.

<span class="kicker">Roadmap</span>

## Further development

- **Collection-level search** (datasets, not millions of granules) for speed and clarity.
- **Caching & smarter ranking** of the most relevant results.
- **Richer visualization**: vertical profiles, time animation of map layers, downloads.
- **A metadata feedback loop** — flag records missing featureType/links to improve the catalogue itself.
- **Wider reach**: point it at other national/European catalogues.

<span class="kicker">Why it matters</span>

## A practical test of FAIR data

- The tool only works because data is **F**indable, **A**ccessible, **I**nteroperable, **R**eusable.
- Where it struggles, it **points directly at metadata we can improve**.
- Open standards + good metadata = data that serves people, not just storage.

If you take one thing away — investing in metadata quality and open standards pays off directly in tools like this. The software is the easy part; the metadata is the asset.
